# Making Unions with Shapefile Polygons

In [ORA](http://cfo.dc.gov/page/office-revenue-analysis), there is a desire to create new neighborhoods for aggregation of economic and tax-related data.  The idea is that we can use these areal aggregations to map trends.  We want these units to have two basic properties:

1. They should capture neighborhoods that straddle major economic corridors in the city; and,
2. They should not be Ward-based (to avoid political association to the extent possible).

The last time I created such an aggregation, I used [ArcMap](http://en.wikipedia.org/wiki/ArcMap).  This is a fine platform, to be sure.  My preference, however, is to learn to do this with open-source tools.  For one, this ensures I can do it again when I no longer have an ESRI license at my disposal.  Secondly, it would better accommodate my workflow since I use Notebooks for almost all of my analysis these days.

In particular, there are two packages of interest:

+ [Shapely](http://toblerity.org/shapely/manual.html) is a package designed for the interaction of objects in space.  In particular, my focus rests on its set-theoretic and constructive methods.  These allow for things like making larger boundaries out of a collection of smaller boundaries and finding the common space captured by overlapping boundaries.
+ [Fiona](http://toblerity.org/fiona/fiona.html) is a Python API for [OGR](http://www.gdal.org/ogr/) the open-source industry standard "simple" feature library.  In a nutshell, it allows for I/O and property modification of spatial files.

So, Fiona provides the wherewithal to read/write files and modify coordinate reference systems.  Shapely enables me to spatially manipulate the features once they are in the namespace.

Gasp!  There is a **MultiPolygon** object lurking.  Ok, full disclosure, I was not exactly prescient on this front.  I spent a decent amount of time trying to figure out why I could not create a **MultiPolygon** object from the full list of neighborhoods.  (I even threw a query out to [SE](http://gis.stackexchange.com/questions/83084/shapely-multipolygon-construction-will-not-accept-the-entire-set-of-polygons).)  After needlessly poring over source code (both for Shapely and [geopandas](https://github.com/kjordahl/geopandas), which has plotting functions I wanted to review), the biggest clue was that I could make a **MultiPolygon** object with 62 neighborhoods, but not 63.

So, I suspect the solution to this is to extract the **MultiPolygon** object, bust it up into its constituent **Polygon** object parts, and append these **Polygon** objects back to the original list.

```python
#Capture MultiPolygon (MP) object
mp_interloper=[x for x in nbhd_list if x.geom_type != 'Polygon'][0]

#Create list of Polygon components
mp_parts=[x for x in mp_interloper]

#Create new names for the parts
mp_parts_names=['DC Village 1','DC Village 2']

#Generate dictionary for new split
mp_parts_dict={'DC Village 1':mp_parts[0],
               'DC Village 2':mp_parts[1]}

#Update list of neighborhoods
nbhd_list2=[x for x in nbhd_list if x.geom_type == 'Polygon'] + mp_parts

#Remove MP from dict
def removekey(d, key):
    r = dict(d)
    del r[key]
    return r

nbhd_dict2=removekey(nbhd_dict,'DC Village')

#Update dictionary
nbhd_dict2.update(mp_parts_dict)

print 'THERE ARE ',len(nbhd_list2),' ASSESSMENT NEIGHBORHOODS IN OUR NEW LIST'
```

Now we can generate our **MP** object....

```python
#Generate MP
nbhds=MultiPolygon(nbhd_list2)

nbhds
```

```python
#Get color map
cm = plt.get_cmap('RdBu')
num_colours = len(nbhds)
 
#Generate plot object
fig = plt.figure(figsize=(16,14))
ax = fig.add_subplot(111)

#Set scope of the plot
minx, miny, maxx, maxy = nbhds.bounds
w, h = maxx - minx, maxy - miny
ax.set_xlim(minx - 0.2 * w, maxx + 0.2 * w)
ax.set_ylim(miny - 0.2 * h, maxy + 0.2 * h)

#Set aspect ratio
ax.set_aspect(1)

#Create container for patches (neighborhoods in this context) and locations
patches = []

#Define corresponding lists for shapely objects and names
nbhd_objects=[x[1] for x in sorted(nbhd_dict2.items())]
nbhd_names=[x[0] for x in sorted(nbhd_dict2.items())]

#For each neighborhood...
for idx, p in enumerate(nbhd_objects):
    #...assign a color based upon index position in the MP object...
    colour = cm(1. * idx / num_colours)
    #...generate a patch representing the neighborhood and throw it in the list
    patches.append(PolygonPatch(p, fc=colour, ec='#555555', lw=0.2, alpha=.4, zorder=1))

#Add the patches to the plot object en masse
ax.add_collection(PatchCollection(patches, match_original=True))

#Remove ticks
ax.set_xticks([])
ax.set_yticks([])

#Set title
plt.title("Shapefile polygons rendered using Shapely")
plt.legend()

#Maximize plot usage by minimizing margins
plt.tight_layout()
```

Now we need to overlay the streets.  The first hurdle is the disparity in datum.  DCGIS shapefiles generally conform to [NAD83](http://en.wikipedia.org/wiki/North_American_Datum), but Basemap does not offer NAD83 as an option.  It does, however, offer the very common [WGS84](http://en.wikipedia.org/wiki/World_Geodetic_System) datum.  In fact, this was the datum we chose above.  We need to convert our shapefiles (both streets and neighborhoods) to the WGS84 datum.  [pyproj](https://code.google.com/p/pyproj/), part of Basemap, provides us with the tools to make these transformations.  Let's start with a simple examples using the coordinate reference systems we care about.

First, we need to read in the street data.

```python
#Open centerline shapefile
streets=fiona.open(data_dir+'StreetCenterlineLn.shp')

#Initiate iterator
st=streets.next()

#Create container for features
st_list=[]

#For each feature...
for st in streets:
    #...throw the feature in the list
    st_list.append(shape(st['geometry']))

streets.meta
```

We can see our source CRS information above.  How many streets are we looking at?

```python
print 'THERE ARE ',len(streets),' IN THE CENTERLINES FILE'
```

Note that we are dealing with a 3D **LineString** object.

```python
print st_list[0].coords[:5]
```

Why is this the case?  DCGIS has provided us with sufficiently rich data to account for road elevation.  For our purposes, however, we don't really care about the third dimension.  Furthermore, this third dimension will actually conflict with Basemap, causing it to throw a `ValueError` exception.  We need to take this into account before transforming the data to a new CRS.

```python
#Generate container for 2D lines
st_list2D=[]

#For each street...
for st in st_list:
    if st.geom_type != 'LineString':
        for st_part in st:
            st_2D=[xy[:2] for xy in st_part.coords]
            st_list2D.append(LineString(st_2D))
    else:
        #...extract the first two dimensions...
        st_2D=[xy[:2] for xy in st.coords]
        #...create new LineString objects and throw them in the list
        st_list2D.append(LineString(st_2D))

st_list2D[0].coords[:5]
```

Basically, we need to roll through each of these **LineString** objects and replace them with **LineString** objects that have had the associated coordinate array transformed into the newly projected coordinates.

```python
#Define source and new projections
source_proj=pyproj.Proj('+proj=lcc +datum=NAD83 +lat_0=37.66666666666666 +lat_1=38.3 +lat_2=39.45 +lon_0=-77.0'+\
                        '+x_0=400000 +y_0=0 +units=m +no_defs +ellps=GRS80 +towgs84=0,0,0')
new_proj=pyproj.Proj('+init=EPSG:4326')
# new_proj=pyproj.Proj('+proj=longlat +datum=WGS84')

#Create container for newly projected LineStrings
st_new_list=[]

#For each street in our 2D list...
for i,st in enumerate(st_list2D):
    #...transform coordinates...
    xx,yy=pyproj.transform(source_proj,new_proj,st_list2D[i].coords.xy[0],st_list2D[i].coords.xy[1])
    #...and create new LineString objects which are then thrown into our list
    st_new_list.append(LineString(zip(xx,yy)))
    
print st_new_list[0].coords[:5]
print st_new_list[1].coords[:5]
```

Ok, let's see what our transformed data looks like.

```python
#Create new plot object
fig,axes=plt.subplots(figsize=(16,18))

#Set aspect ratio
axes.set_aspect(1.2)

#Define simple functions to write lines
def plot_line(ax, ob):
    x, y = ob.xy
    ax.plot(x, y, color='#999999', linewidth=.5, solid_capstyle='round', zorder=1)

#For each line in our list...
for st in st_new_list:
    #...plot the line
    plot_line(axes,st)
    
#Remove ticks
axes.set_xticks([])
axes.set_yticks([])

#Set title
plt.title("Street Centerlines in DC")
plt.legend()

#Maximize plot usage by minimizing margins
plt.tight_layout()
```

Let's read in our assessment neighborhoods as a geopandas GeoDataFrame (GDF) and transform the coordinates to WGS84.

```python
#Read in data
aneigh=gp.GeoDataFrame.from_file(data_dir+'ASNBHDPLY.shp')

#Transform to WGS84
aneigh_ll=aneigh.to_crs(crs={'proj':'longlat',
                             'datum':'WGS84'}).rename(columns={'geometry':'geom_temp'})

aneigh_ll.head()
```

The coordinates are backwards (an ESRI convention), so we need to switch the order.  Note that one neighborhood, DC Village, is still split in two.  Our earlier mapping exercise reveals that it's not a big deal to drop the second component for our purposes (it would not alter the decision to include DC Village in one neighborhood cluster versus another).  For parsimony, the second component is just dropped below.

```python
#Define function to swap lat-long coords in Polygon objects
def swap_coord(poly):
    '''Mappable function that switches coordinates in polygon objects'''
    #If the object is not a Polygon...
    if poly.geom_type != 'Polygon':
        #...take the first component and switch the coordinates
        new_poly=Polygon([(x[1],x[0]) for x in poly[0].exterior.coords])
    #...otherwise...
    else:
        #...swap the coordinates of the Polygon
        new_poly=Polygon([(x[1],x[0]) for x in poly.exterior.coords])
    return new_poly

#Swap coordinates
aneigh_ll['geometry']=aneigh_ll['geom_temp'].apply(lambda x: swap_coord(x))

#Drop geom_temp
an_ll=aneigh_ll[[x for x in aneigh_ll.columns if x != 'geom_temp']]
        
an_ll.head()
```

```python
#Write transformed data to disk
an_ll.to_file('asmnt_neigh_ll2.json',driver='GeoJSON')

#Read it back in as a fiona collection object
# an_ll2=fiona.open('asmnt_neigh_ll.shp')

# an_ll2
```
