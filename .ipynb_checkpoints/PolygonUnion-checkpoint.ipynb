# Making Unions with Shapefile Polygons

In [ORA](http://cfo.dc.gov/page/office-revenue-analysis), there is a desire to create new neighborhoods for aggregation of economic and tax-related data.  The idea is that we can use these areal aggregations to map trends.  We want these units to have two basic properties:

1. They should capture neighborhoods that straddle major economic corridors in the city; and,
2. They should not be Ward-based (to avoid political association to the extent possible).

The last time I created such an aggregation, I used [ArcMap](http://en.wikipedia.org/wiki/ArcMap).  This is a fine platform, to be sure.  My preference, however, is to learn to do this with open-source tools.  For one, this ensures I can do it again when I no longer have an ESRI license at my disposal.  Secondly, it would better accommodate my workflow since I use Notebooks for almost all of my analysis these days.

In particular, there are two packages of interest:

+ [Shapely](http://toblerity.org/shapely/manual.html) is a package designed for the interaction of objects in space.  In particular, my focus rests on its set-theoretic and constructive methods.  These allow for things like making larger boundaries out of a collection of smaller boundaries and finding the common space captured by overlapping boundaries.
+ [Fiona](http://toblerity.org/fiona/fiona.html) is a Python API for [OGR](http://www.gdal.org/ogr/) the open-source industry standard "simple" feature library.  In a nutshell, it allows for I/O and property modification of spatial files.

So, Fiona provides the wherewithal to read/write files and modify coordinate reference systems.  Shapely enables me to spatially manipulate the features once they are in the namespace.

After needlessly poring over source code (both for Shapely and [geopandas](https://github.com/kjordahl/geopandas), which has plotting functions I wanted to review), the biggest clue was that I could make a **MultiPolygon** object with 62 neighborhoods, but not 63.

So, I suspect the solution to this is to extract the **MultiPolygon** object, bust it up into its constituent **Polygon** object parts, and append these **Polygon** objects back to the original list.

```python
#Capture MultiPolygon (MP) object
mp_interloper=[x for x in nbhd_list if x.geom_type != 'Polygon'][0]

#Create list of Polygon components
mp_parts=[x for x in mp_interloper]

#Create new names for the parts
mp_parts_names=['DC Village 1','DC Village 2']

#Generate dictionary for new split
mp_parts_dict={'DC Village 1':mp_parts[0],
               'DC Village 2':mp_parts[1]}

#Update list of neighborhoods
nbhd_list2=[x for x in nbhd_list if x.geom_type == 'Polygon'] + mp_parts

#Remove MP from dict
def removekey(d, key):
    r = dict(d)
    del r[key]
    return r

nbhd_dict2=removekey(nbhd_dict,'DC Village')

#Update dictionary
nbhd_dict2.update(mp_parts_dict)

print 'THERE ARE ',len(nbhd_list2),' ASSESSMENT NEIGHBORHOODS IN OUR NEW LIST'
```

Why is this the case?  DCGIS has provided us with sufficiently rich data to account for road elevation.  For our purposes, however, we don't really care about the third dimension.  Furthermore, this third dimension will actually conflict with [Basemap](http://matplotlib.org/basemap/) (if we went that route), causing it to throw a `ValueError` exception.  We need to take this into account before transforming the data to a new CRS.

```python
#Generate container for 2D lines
st_list2D=[]
st_name_list2D=[]

#For each street...
for i,st in enumerate(st_list):
    #...if the shape is not a LineString...
    if st.geom_type != 'LineString':
        #...break it up and add each individually...
        for st_part in st:
            st_2D=[xy[:2] for xy in st_part.coords]
            st_list2D.append(LineString(st_2D))
            st_name_list2D.append(st_name_list[i])
    else:
        #...extract the first two dimensions...
        st_2D=[xy[:2] for xy in st.coords]
        #...create new LineString objects and throw them in the list
        st_list2D.append(LineString(st_2D))
        st_name_list2D.append(st_name_list[i])

st_list2D[0].coords[:5]
```

We can capture the segments for each road in two summary lists: `corridors` and `major_roads`.  Let us write a function that pairs names with the associated `LineString` segments.

```python
corridor_labels=['Massachusetts Ave','Connecticut Ave','16th St','New York Ave','Pennsylvania Ave']
major_road_labels=['Georgia Ave','North Capitol St','Rhode Island Ave','East Capitol St','South Capitol St']

def street_capture(street_names):
    #Create container for all streets in the group
    all_streets=[]
    #For each street name...
    for street in street_names:
        #...capture the index position...
        idx=[x[0] for x in enumerate(st_name_list2D) if street in str(x[1])]
        #...pair street names and segments based upon index position...
        st_pair=[(street,x[1]) for x in enumerate(st_list2D) if x[0] in idx]
        #...and for each pair in st_pair
        for pair in st_pair:
            #...throw said pair in the list
            all_streets.append(pair)
    return all_streets

#Capture all corridors and major roads
corridors=street_capture(corridor_labels)
major_roads=street_capture(major_road_labels)
```

```python
#Define group lists
mass1=['Palisades','Kent','Fort Drive','Berkley','Foxhall','Spring Valley','Wesley Heights','Glover - Archbold Parkway',\
       'Glover Park','Burleith','Georgetown','Observatory Circle','American University','Cleveland Park',\
       'Massachusetts  Avenue Heights','Rock Creek Park']
conn1=['Chevy Chase','Hawthorne','Wakefield','North Cleveland Park','Woodley','Garfield','Kalorama','Forest Hills',\
       'National Zoological Park']
sixt1=['Colonial Village','Crestwood','Mt. Pleasant','Shepherd Park','Brightwood','16th Street Heights','Columbia Heights',\
       'Ledroit Park','Takoma Park','Chillum','Petworth','Brookland','Riggs Park']
ny1=['Eckington','Brentwood','Michigan Park','Woodridge','National Arboretum','Trinidad','Fort Lincoln']
penn1=['DC Stadium Area','Lily Ponds','Deanwood','Fort Dupont Park','Marshall Heights','Hillcrest','Anacostia Park',\
       'Anacostia','Barry Farms','Randle Heights',"St. Elizabeth's Hospital",'Bolling Air Force Base','Congress Heights',\
       'DC Village 1','DC Village 2']
cenW1=['Foggy Bottom','Central-tri 1']
cenN1=['Old City 2','Central-tri 3','R.L.A. (N.E.)']
cenSE1=['Mall/East Potomac Park','R.L.A. (S.W.)','Capitol Hill','Old City 1','Ft. McNair','Washington Navy Yard']

#Generate list of groups
group1=[mass1,conn1,sixt1,ny1,penn1,cenW1,cenN1,cenSE1]

#Define function to map neighborhood groups to colors
def nbhd_col_map(grp,color_map):
    '''Function takes a list of neighborhood groupings and returns a dict of color pairings by group.'''
    #Set list of colors
    nbhd_colors=color.colors_from_cmap(len(grp),color_map)
    #Create a container to pair neighborhoods and colors
    nbhd_color_map=[]
    #For each neighborhood list...
    for i,nbhd_list in enumerate(grp):
        #...and each neighborhood in the list...
        for nbhd in nbhd_list:
            #...pair the neighborhood to the neighborhood-list specific color and throw it in the list
            nbhd_color_map.append((nbhd,nbhd_colors[i]))      
    #Create color dictionary
    nbhd_color_dict=dict(nbhd_color_map)
    return nbhd_color_dict

nbhd_col_map(group1,'Dark2')
```

Now that we have colors mapped, we can look at the aggregate groupings.  Since different groupings will be tested, let's create a function.

```python
def plot_neighborhoods(group_list,title):
    #Generate plot object
    fig = plt.figure(figsize=(20,20))
    ax = fig.add_subplot(111)
    
    #Set scope of the plot
    minx, miny, maxx, maxy = nbhds.bounds
    w, h = maxx - minx, maxy - miny
    ax.set_xlim(minx - 0.2 * w, maxx + 0.2 * w)
    ax.set_ylim(miny - 0.2 * h, maxy + 0.2 * h)
    
    #Set aspect ratio
    ax.set_aspect(1)
    
    #Create container for patches (neighborhoods in this context) and locations
    patches = []
    patch_loc=[]
    
    #For each neighborhood...
    for idx, p in enumerate(nbhd_objects):
        #...assign a color based upon index position in the MP object...
        colour = nbhd_col_map(group_list,'Dark2')[nbhd_names[idx]]
        #...generate a patch representing the neighborhood and throw it in the list...
        patches.append(PolygonPatch(p, fc=colour, ec='#555555', lw=0.2, alpha=.2, zorder=1,label=nbhd_names[idx]))
        #...and capture the patch location
        patch_loc.append((p.centroid.x-250,p.centroid.y))
    
    #Add the patches to the plot object en masse
    ax.add_collection(PatchCollection(patches, match_original=True))
    
    #Add labels
    for i,loc in enumerate(patch_loc):
        ax.annotate(nbhd_names[i],xy=loc,size=5)
        
    #For each line in our list...
    for st in corridors:
        #...plot the line
        plot_line(ax,st[1],col='k',lw=4)
        
    #For each line in our list...
    for st in major_roads:
        #...plot the line
        plot_line(ax,st[1],lw=3)
        
    #Add road labels
    ax.annotate('Massachusetts Ave',xy=(390500,143000),size=8,rotation=45)
    ax.annotate('Connecticut Ave',xy=(392600,145000),size=8,rotation=45)
    ax.annotate('16th St',xy=(396600,147500),size=8,rotation=-45)
    ax.annotate('Georgia Ave',xy=(397400,146700),size=8,rotation=-45)
    ax.annotate('N.Capitol St',xy=(399000,145200),size=8,rotation=-45)
    ax.annotate('Rhode Island Ave',xy=(402800,141400),size=8,rotation=-45)
    ax.annotate('New York Ave',xy=(404800,139400),size=8,rotation=-45)
    ax.annotate('E.Capitol St',xy=(407400,135900),size=8,rotation=45)
    ax.annotate('Pennsylvania Ave',xy=(404300,133200),size=8,rotation=45)
    ax.annotate('S.Capitol St',xy=(399700,128200),size=8,rotation=45)
    
    #Remove ticks
    ax.set_xticks([])
    ax.set_yticks([])
    
    #Set title
    plt.title(title)
    plt.legend()
    
    #Maximize plot usage by minimizing margins
    plt.tight_layout()

plot_neighborhoods(group1,'First Grouping')
plt.savefig('GROUP1_Corridors.png',dpi=200)
```
