# 5-Year ACS Data for the District (2005-2009)

Demographic information is a critical piece of the neighborhood report I am currently constructing for the DC Office of Revenue Analysis.  The neighborhoods we are using are built from [Census Block Groups](http://en.wikipedia.org/wiki/Census_block_group), aggregated in a way that roughly corresponds to the planning areas identified by the [DC Office of Planning](http://planning.dc.gov/DC/Planning) in the [comprehensive plan](http://planning.dc.gov/DC/Planning/Across+the+City/Comprehensive+Plan).  The first step, before capturing trends, is to understand the current distribution of socioeconomic activity to the extent possible.  The most recent data on this activity is collected by the [American Community Survey](http://www.census.gov/acs/www/).  Block group level data is not available for the 1-year and 3-year data collections, but it is available for the 5-year collection.  This script roughly follows the protocol established in [DC_ACS5yr_08_12](http://127.0.0.1:8879/6c308532-2511-4be5-b5e3-53bb1d913e54#5-Year-ACS-Data-for-the-District-(2008-2012)) for generating a dataset with demographic information that can be binded to Block Groups.

## Sequence Lookup

The first step in the process is to grab the sequence lookup table and capture its contents in a DataFrame.

```python
#Establish working directory
workdir='/home/choct155/ora_public/acs/yr05_09/'

#Read in lookup info
seq_temp=pd.read_csv(workdir+'Sequence_Number_and_Table_Number_Lookup.csv',dtype={'Sequence Number':str})

#Forward fill missing values
seq=seq_temp.fillna(method='ffill')

#Replace column names
seq.columns=['file','table','seq_temp','line','start','tot_cells_table','tot_cells_seq','title','subject']

#Generate leading zeros for sequence number
seq['seq']=seq['seq_temp'].apply(lambda x: '{0:0>4}'.format(x))

seq.head()
```

Now we can categorize these files by sequence ID.  Fortunately, the Census Bureau prizes consistency in file naming convention.

```python
#Capture list of files
files=[x.split('/')[6] for x in glob.glob(workdir+'*')]

#Capture estimate files
est_files=sorted([x for x in files if x.startswith('e')])

#Capture margin of error files
moe_files=sorted([x for x in files if x.startswith('m')])

#Create DataFrame capturing subject associations by file
f_sub_df=DataFrame({'est':est_files,
                   'moe':moe_files})

#Check correspondence across estimate and margin of error files
f_sub_df['match']=f_sub_df['est'].str[1:]==f_sub_df['moe'].str[1:]
print f_sub_df['match'].value_counts()

#Capture list of tuples paring sequence ID and subject
seq_sub_pairs=zip(seq['seq'],seq['subject'])

#Create dictionary from pairings
seq_sub_dict=dict(seq_sub_pairs)

#Create subject area variable
f_sub_df['sub']=f_sub_df['est'].apply(lambda x: seq_sub_dict[x[8:12]])

#Write to disk
# f_sub_df.to_csv(workdir+'subject_map05_09.csv')

f_sub_df.head(10)
```

## Raw Data Estimates

Let's take a quick look at the `SEX BY AGE` and add a count variable for each row within a given `table`.

```python
#Create count variable
seq['table_count']=seq.groupby('table').cumcount()

seq.ix[1418:1425]
```

Six columns in the raw data contain identifying information:

    ['FILEID','FILETYPE','STUSAB','CHARITER','SEQUENCE','LOGRECNO']
    
The seventh column begins the raw data component.  For the *Age & Sex* data, this corresponds with the 1422nd row of the lookup table (where `line==1`).  Note also that the universe in question is defined in the preceding row (`title == Universe:  Total Population`).  This is a consistent pattern for each table, and thus the pattern repeats within sequence.  

We need a generic method of pulling all of the relevant column names for each file, prepending the standard identifiers, and using them as column names in our set.  If the structure was flat, we could just grab the lists directly.  However, the labels nest within sub-universe, which nest within table, which nest within sequence.  Furthermore, our column space in the output set will need to be hierarchical to correctly both the label and the universe definitions.  

The positioning of the universe definition (and its components) is a bit of an issue because it hinders clean gathering of the information in that column.  Unlike the broad *SEX BY AGE* classifier, universe and subuniverse are critical pieces of the hierarchy, and thus cannot be subsetted out. One solution is to just kick universe and sub-universe out into their own variables. Once we have done this, we can create a column of tuples that captures universe, sub-universe, and group.

```python
#Create universe variable
seq['universe']=Series([x[11:] if 'Universe' in x else NaN for x in seq['title']],index=seq.index).\
                    fillna(method='ffill')
    
#Create sub-universe variable
seq['subuniv']=Series([x if any(':' in x or x=='Total') else NaN for x in seq['title']],index=seq.index).str.\
                replace(':','').fillna(method='ffill')
    
#Create clean group variable
seq['group']=seq['title'].str.replace(':','')

#Remove broad, but redundant categories and newly replaced universe titles
seq2=seq[seq['table_count']>1].set_index('seq')
    
#Create tuple for column name
seq2['colnames']=seq2.apply(lambda x: (x['universe'],x['subuniv'],x['group']),axis=1)

seq2.ix['0007'].head()
```

Now let's read in the data to see how these names match up.  Getting the hierarchical information from the universe/sub-universe/group hierarchy to mesh with the single level identifier variables is actually a less than straightforward process.  We want the output dataset to have four dimensions:  universe, sub-universe, group, and logical record number.  The data should be arranged in six columns:  file ID, file type, state ID, iteration number, sequence number, and the actual data value.

To make this happen, we need to divorce the identifying information from the data, and manipulate both components so that we may join the two on the common key:  `LOGRECNO`. In DC_ACS5yr_08_12 we defined a function that takes the filename as an input.

```python
#Create list of ID variables
idcols=['FILEID','FILETYPE','STUSAB','CHARITER','SEQUENCE','LOGRECNO']

def id2data(f,file_loc=workdir,lkup=seq2,id_list=idcols):
    '''Function joins raw data with identifying information and returns a DataFrame that permits slicing
    by universe, sub-universe, and group'''
    #Read in data
    file_all=pd.read_csv(file_loc+f,header=None,dtype={3:str,
                                                       4:str})
    
    #Subset to ID information
    ids=file_all[file_all.columns[:6]]
    
    #Rename columns to ID fields
    ids.columns=id_list
    
    #Set logical record number to the index
    ids.set_index('LOGRECNO',inplace=True)
    
    #Capture the data component of the file
    data=file_all[file_all.columns[5:]].rename(columns={5:'LOGRECNO'})
    
    #Set logical record number to the index
    data.set_index('LOGRECNO',inplace=True)
    
    #Extract sequence number
    seq_no=f[8:12]
#     print list(lkup.ix[seq_no]['colnames'])
    #Replace columns with hierarchical column index
    data.columns=pd.MultiIndex.from_tuples(list(lkup.ix[seq_no]['colnames']),names=['universe','subuniv','group'])
    
    #Stack the columns into a hierarchical index
    id_data=DataFrame(data.stack(['universe','subuniv','group'])).rename(columns={0:'VALUE'})
    
    #For each ID column...
    for col in [x for x in id_list if x != 'LOGRECNO']:
        #...merge ID info with the data DF on logical record number
        id_data[col]=ids[col].ix[id_data.index.get_level_values(level='LOGRECNO')].values
        
    return id_data

id2data('e20095dc0016000.txt').head()
```
