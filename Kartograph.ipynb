# [Kartograph](http://kartograph.org/docs/kartograph.py/)

Kartograph is a mapping package that seeks to provide the best of two worlds.  It provides fine aesthetic control in the generation of SVG maps via Kartograph.py, and interactive capability/web presentation via Kartograph.js (potentially using Kartograph.py output as the starting point).  In this Notebook we will explore this capability further.

```python
from kartograph import Kartograph
from IPython.display import SVG
import geopandas as gp
import fiona
from shapely.geometry import mapping, shape, Polygon
import pandas as pd
from collections import OrderedDict
```

## Basic Map

The logical starting point is just to see if we can create an image of a boundary set that we care about.  Given the ongoing development of the neighborhood trends report in DC, the shapefile I have constructed for this purpose (see *PolygonUnion*) seems a logical starting point.  The basic process for this has three steps:

1. Instantiate a **`Kartograph`** object;
2. Craft a [JSON](http://www.json.org/) tree with the relevant parameters (inclusion layers, filtering, aesthetics, etc.); and,
3. Process the JSON tree with **`K.generat()`** and output the SVG image.

This same basic pattern will persist throughout all map generating processes.  To start, we will simply add a layer from a shapefile.

**NOTE**:  It is important to keep in mind that Kartograph is still under development (like any tool).  It does not seem, at the current time, to handle [Lambert Conformal Conic](http://en.wikipedia.org/wiki/Lambert_conformal_conic_projection) projections well.  When a shapefile with this projection is encountered, an exception is thrown when the map is being generated:

    RuntimeError: projection not named
    
This exception occurs even if we try to transform the data to a different projection within the JSON tree.  This is a particular problem for DC data, which tends to be LCC.  The workaround, at this point, is to convert to a different projection prior to processing the data with Kartograph.  See *ProjectionConvert* for the workflow on this task.

```python
#Instantiate K object
K = Kartograph()

#Define JSON tree
nbhds={"layers": [{"id":"mylayer",
                   "src": "/home/choct155/ora_public/dc_shp/Google_CensusBG_Areas.shp",
                   "attributes":["NAME"],
                   "labeling":{"key":"NAME"}
                   }],
       "proj":{"id":"mercator"
               },
       "export": {"width": 500,
                  "ratio": 0.5
                  }
       }

#Generate map
K.generate(nbhds, outfile='DC_nbhds.svg')
```

Let us observe the fruits of our labor.

```python
SVG('DC_nbhds.svg')
```

## CSS Styling

```python
css='#mylayer {fill:steelblue;'+\
               'stroke:white},'+\
    '#mylayer-label{font-family:Serif;'+\
                   'font-size:8px;'+\
                   'fill:white}'

K.generate(nbhds, outfile='DC_nbhds.svg',stylesheet=css)
```

## Data Integration

So, what happens if we have our map here, and we want to make a chloropleth?  Our current shapefile does not contain any usable data (just names).  How flexible can we be if we read in the shapefile in question, and join some data to it?

```python
#Set file locations
shp_dir='/home/choct155/ora_public/dc_shp/'
dem_dir='/home/choct155/ora_public/acs/yr08_12/'

#Read in geo data
nbhd_shp=gp.GeoDataFrame.from_file(shp_dir+'Google_CensusBG_Areas.shp')

#Read in demographic data
dem=pd.read_csv(dem_dir+'DC_age_by_area.csv')

#Join the demographic data to the shapefile
data=nbhd_shp.join(dem)

#Create more manageable names
display_names=['Rock Creek (E)','Rock Creek (W)','Upper\nNortheast','Mid-City','Near\nNorthwest','SW Waterfront',\
               'Far East','Central\nWashington','Capitol\nHill','Far South']
display_dict=dict(zip(list(data['NAME']),display_names))

#Map display labels into data
data['labels']=data['NAME'].map(display_dict)

data
```
