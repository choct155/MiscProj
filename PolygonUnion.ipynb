# Making Unions with Shapefile Polygons

In [ORA](http://cfo.dc.gov/page/office-revenue-analysis), there is a desire to create new neighborhoods for aggregation of economic and tax-related data.  The idea is that we can use these areal aggregations to map trends.  We want these units to have two basic properties:

1. They should capture neighborhoods that straddle major economic corridors in the city; and,
2. They should not be Ward-based (to avoid political association to the extent possible).

The last time I created such an aggregation, I used [ArcMap](http://en.wikipedia.org/wiki/ArcMap).  This is a fine platform, to be sure.  My preference, however, is to learn to do this with open-source tools.  For one, this ensures I can do it again when I no longer have an ESRI license at my disposal.  Secondly, it would better accommodate my workflow since I use Notebooks for almost all of my analysis these days.

In particular, there are two packages of interest:

+ [Shapely](http://toblerity.org/shapely/manual.html) is a package designed for the interaction of objects in space.  In particular, my focus rests on its set-theoretic and constructive methods.  These allow for things like making larger boundaries out of a collection of smaller boundaries and finding the common space captured by overlapping boundaries.
+ [Fiona](http://toblerity.org/fiona/fiona.html) is a Python API for [OGR](http://www.gdal.org/ogr/) the open-source industry standard "simple" feature library.  In a nutshell, it allows for I/O and property modification of spatial files.

So, Fiona provides the wherewithal to read/write files and modify coordinate reference systems.  Shapely enables me to spatially manipulate the features once they are in the namespace.

```python
import numpy as np
import fiona 
from shapely.geometry import shape, MultiPolygon
from descartes import PolygonPatch
from matplotlib.collections import PatchCollection
import geopandas as gp
from mpl_toolkits.basemap import Basemap
```

The goal is to create eight regions in the District, give or take.  This script will serve as the basis for constructing various alternative schemes.  To start with, we will use [assessment neighborhoods](http://dcatlas.dcgis.dc.gov/metadata/AsNbhdPly.html) as the primary building blocks.  If we find we need more resolution, perhaps we can employ smaller [Census boundaries](http://www.census.gov/geo/maps-data/data/tiger.html).  

First, we need to read in the data.

```python
#Establish data directory
data_dir='/home/choct155/dissertation/MiscData/'

#Open assessment neighborhood shapefile
asmt=fiona.open(data_dir+'ASNBHDPLY.shp')

#Initiate iterator
nbhd=asmt.next()

#Create container for features and names
nbhd_list=[]
nbhd_name_list=[]

#For each feature...
for nbhd in asmt:
    #...throw the feature in the list...
    nbhd_list.append(shape(nbhd['geometry']))
    #...and its name into another list
    nbhd_name_list.append(nbhd['properties']['NEIGHBORHO'])

asmt
```

It appears that opening a shapefile in Fiona is quite analagous to opening any file with base Python.  Before doing anything else, let's map names to each neighborhood.

```python
nbhd_dict=dict(zip(nbhd_name_list,nbhd_list))
```

It is simpy Shapely object that features methods for set-theoretic and constructive operations.  It also has introspective methods like `geom_type` and `centroid` shown above.  To see the coordinate representation of all or part of the object, we can employ the `wkt` extension.

First, however, we need to generate a **MultiPolygon** object.

Before we ge there, a little validation is in order.  To work with these data visually, it's useful to verify that we are operating with the data types we think we are.  Are there any geometries that are not **Polygon** objects.

```python
[x for x in nbhd_dict.items() if x[1].geom_type != 'Polygon']
```

Gasp!  There is a **MultiPolygon** object lurking.  Ok, full disclosure, I was not exactly prescient on this front.  I spent a decent amount of time trying to figure out why I could not create a **MultiPolygon** object from the full list of neighborhoods.  (I even threw a query out to [SE](http://gis.stackexchange.com/questions/83084/shapely-multipolygon-construction-will-not-accept-the-entire-set-of-polygons).)  After needlessly poring over source code (both for Shapely and [geopandas](https://github.com/kjordahl/geopandas), which has plotting functions I wanted to review), the biggest clue was that I could make a **MultiPolygon** object with 62 neighborhoods, but not 63.

So, I suspect the solution to this is to extract the **MultiPolygon** object, bust it up into its constituent **Polygon** object parts, and append these **Polygon** objects back to the original list.

```python
#Capture MultiPolygon (MP) object
mp_interloper=[x for x in nbhd_list if x.geom_type != 'Polygon'][0]

#Create list of Polygon components
mp_parts=[x for x in mp_interloper]

#Create new names for the parts
mp_parts_names=['DC Village 1','DC Village 2']

#Generate dictionary for new split
mp_parts_dict={'DC Village 1':mp_parts[0],
               'DC Village 2':mp_parts[1]}

#Update list of neighborhoods
nbhd_list2=[x for x in nbhd_list if x.geom_type == 'Polygon'] + mp_parts

#Remove MP from dict
def removekey(d, key):
    r = dict(d)
    del r[key]
    return r

nbhd_dict2=removekey(nbhd_dict,'DC Village')

#Update dictionary
nbhd_dict2.update(mp_parts_dict)

print 'THERE ARE ',len(nbhd_list2),' ASSESSMENT NEIGHBORHOODS IN OUR NEW LIST'
```

Now we can generate our **MP** object....

```python
#Generate MP
nbhds=MultiPolygon(nbhd_list2)

nbhds
```

...and plot our neighborhoods.

```python
#Get color map
cm = plt.get_cmap('RdBu')
num_colours = len(nbhds)
 
#Generate plot object
fig = plt.figure(figsize=(16,14))
ax = fig.add_subplot(111)

#Set scope of the plot
minx, miny, maxx, maxy = nbhds.bounds
w, h = maxx - minx, maxy - miny
ax.set_xlim(minx - 0.2 * w, maxx + 0.2 * w)
ax.set_ylim(miny - 0.2 * h, maxy + 0.2 * h)

#Set aspect ratio
ax.set_aspect(1)

#Create container for patches (neighborhoods in this context) and locations
patches = []
patch_loc=[]

#Define corresponding lists for shapely objects and names
nbhd_objects=[x[1] for x in sorted(nbhd_dict2.items())]
nbhd_names=[x[0] for x in sorted(nbhd_dict2.items())]

#For each neighborhood...
for idx, p in enumerate(nbhd_objects):
    #...assign a color based upon index position in the MP object...
    colour = cm(1. * idx / num_colours)
    #...generate a patch representing the neighborhood and throw it in the list...
    patches.append(PolygonPatch(p, fc=colour, ec='#555555', lw=0.2, alpha=.4, zorder=1,label=nbhd_names[idx]))
    #...and capture the patch location
    patch_loc.append((p.centroid.x,p.centroid.y))

#Add the patches to the plot object en masse
ax.add_collection(PatchCollection(patches, match_original=True))

#Remove ticks
ax.set_xticks([])
ax.set_yticks([])

#Set title
plt.title("Shapefile polygons rendered using Shapely")
plt.legend()

#Maximize plot usage by minimizing margins
plt.tight_layout()
```
