# Spatial Association and Polygon Unions

In *MapLayering* we used [fiona](http://toblerity.org/fiona/), [shapely](http://toblerity.org/shapely/manual.html), and [descartes](https://pypi.python.org/pypi/descartes) to help us select [assessment neighborhoods](http://data.dc.gov/Metadata.aspx?id=127) and [Census block groups](http://www.census.gov/geo/reference/gtc/gtc_bg.html) we would like to aggregate into larger sub-District geographies.  The maps generated in that script and [one](https://choct155.cartodb.com/viz/e1100d60-8923-11e3-a59c-04010791e701/map) built with [CartoDB](http://cartodb.com/) helped ORA staff get a sense of how many neighborhoods were required and what the boudaries should generally look like.  The consensus was to utilize Census block groups to match the boundaries of the planning areas identified by the [Office of Planning](http://planning.dc.gov/DC/Planning) in the [comprehensive plan](http://dc.gov/DC/Planning/Across+the+City/Comprehensive+Plan) (OP).  This script is designed to capture the block group collections that most closely match the areas identified by OP.  

```python
#Basic Data Manipulation
import numpy as np
import pandas
from pandas import Series,DataFrame

#Spatial I/O
import fiona
from fiona import collection

#Spatial Geometry Manipulation
from shapely.geometry import mapping,shape,Point,Polygon,MultiPolygon
from shapely.ops import unary_union
from descartes import PolygonPatch
from rtree import index

#Visualization
from matplotlib.collections import PatchCollection
from mpltools import color
```

While we could have simply used the area boundaries, we have opted to use block groups in order to leverage the associated demographic information collected by the Census Bureau.  The complication that arises is inconsistency in the boundaries.  The block groups do not nest perfectly into planning areas.  There are two approaches we could take to remedy this issue:

1. Develop a rule to proportionally allocate data from block groups that are split by a planning area boundary.  For example, if 60% of the block group fell within planning area $x$, $x$ would receive 60% of the value of said block group.
2. Use the planning areas as only a guide, and do not modify the block group values.

The first approach is useful if we need strict adherence to planning boundaries, but it imposes a strong assumption about the distribution of a given attribute.  For example, if we are trying to allocate people across planning areas, our proportional allocation rule assumes uniform distribution of population within the block group.  This is a reasonable assumption *in the absence of additional information*, but it is likely to be inaccurate.  For our purposes, however, strict adherence to the planning boundaries is not essential.  We can, thus, proceed with option \#2 and preserve the integrity of the Census data.

With respect to how we will identify which block groups we can associate with a given planning area, a reasonable standard is a plurality rule:  block groups will be associated with the planning area that contains the largest share of the block group.  This is a plurality rule (as opposed to majority) because it is conceivable that block groups are split by more than two areas.  Calculating proportional allocations for each block group is one way to pursue adherence to this rule, but it seems it a bit cumbersome for this task.  A more lightweight approach would be to identify the intersection of planning areas and block group [centroids](http://en.wikipedia.org/wiki/Centroid).  In the absence of convexity, the centroid provides an average position for a given block group.  Therefore, if a planning area contains the centroid, it contains the majority of the block group by construction.  Convexity is a fairly limited issue with block groups in the District, so this approximation should be of sufficient quality.

## Data I/O

The first step is to read in the block group and planning area shapefiles.  First the block groups...

```python
#Establish data directories
data_dir='/home/choct155/dissertation/MiscData/'
dc_dir='/home/choct155/ora_public/dc_shp/'

#Open assessment neighborhood shapefile
bgroups=fiona.open(data_dir+'BlockGroupPly.shp')

#Initiate iterator
bgrp=bgroups.next()

#Create container for features and names
bg_list=[]
bg_id_list=[]

#For each feature...
for bg in bgroups:
    #...throw the feature in the list...
    bg_list.append(shape(bg['geometry']))
    #...and its name into another list
    bg_id_list.append(bg['properties']['GEOID'])
    
#Create dictionary mapping IDs and polygons
bg_dict=dict(zip(bg_id_list,bg_list))

bgroups.meta
```

..and then the planning areas.

```python
#Open assessment neighborhood shapefile
pareas=fiona.open(dc_dir+'CompPlanPlanningAreas.shp')

#Initiate iterator
pa=pareas.next()

#Create container for features,IDs, and names
pa_list=[]
pa_id_list
pa_name_list=[]

#For each feature...
for area in pareas:
    #...throw the feature in the list...
    pa_list.append(shape(area['geometry']))
    #...and its name and ID into other lists
    pa_id_list.append(area['properties']['OBJECTID'])
    pa_name_list.append(area['properties']['NAME'])

#Create mapping of area OBJECTIDs and polygons
pa_dict=dict(zip(pa_id_list,pa_list))

pareas.meta
```

## Capturing Block Group Planning Area Intersection by Centroid

Fortunately, matters are simplified by the common [CRS](http://en.wikipedia.org/wiki/Spatial_reference_system) across the sets.  We have our list of planning polygons, but we still need to extract the centroids from the block groups.

```python
#Capture centroids from each block group poly
bg_centroids=[(bg.centroid.x,bg.centroid.y) for bg in bg_list]
bg_centroids[:5]
```

To link the centroid points to the planning areas in which they are contained, we will employ another leg of the [Toblerity](http://toblerity.org/) project:  [Rtree](http://toblerity.org/rtree/).  Rtree is package that "provides a number of advanced spatial indexing features for the spatially curious Python user."  For our purposes, it has a very useful intersection function.  The first step is creating a spatial index for the polygons.  We are using unique identifiers (`GEOID`) for each planning area's [bounding box](http://en.wikipedia.org/wiki/Minimum_bounding_box) to populate this index.  

```python
#Instantiate index object
idx=index.Index()

#For each planning area...
for i,parea in enumerate(pa_list):
    #...assign index value to each area
    idx.insert(int(pa_id_list[i]),parea.bounds)
```

Now that we have this index, we can assign centroids to each of the index elements based upon spatial intersection.  To make this happen, we will employ the `intersection()` method for the **`index.Index()`** class.  Each intersection operation returns a generator, telling us which block group contains the point in question.  For example...

```python
idx.intersection(bg_centroids[0]).next()
```

...the first considered centroid is contained within the fifth planning area.  We need to perform the same mapping for all points in our list of centroids.  A few subtleties occur below:

1. The output of the `intersection()` method is an integer, not a **`Polygon`**;
2. The input for the `intersection()` method is a coordinate pair, nothe **`Point`** object itself; and,
3. The inputs for the `within()` method are both geometric objects (in this instance **`Point`** & **`Polygon`**).

```python
#Create container for area matches
amatches=[]

#For each pair of centroid coordinates...
for cpt in bg_centroids:
    #...instantiate a temporary match object...
    temp_match=None
    #...and for each bounding box that contains the centroid...
    for area in idx.intersection(cpt):
        #...if centroid is in the polygon (not just the bounding box)...
        if Point(cpt).within(pa_dict[area]):
            #...update the temporary match object...
            temp_match=area
            break
    #...and throw it in the list
    amatches.append(temp_match)
```

Now we have a vector of planning area identifiers that are linked to individual centroids and, in turn, individual block groups.

## Visual Inspection of Groupings

Before we try to join the block groups into our broader neighborhoods, it would be prudent to inspect the mapping we have created to ensure that we are getting what we believe are appropriate groupings.  Below we construct a map which colors block groups by their associated planning area (identified in the above intersection).  The planning areas boundaries are overlaid so that we can see how our block group collections differ from the planning areas at the margin.

```python
#Get color map
cm=plt.get_cmap('Set1')
num_colors=len(pa_list)

#Generate plot object
fig,ax=plt.subplots(figsize=(16,14))

#Set scope of plot
minx,miny,maxx,maxy=MultiPolygon(pa_list).bounds
w,h=maxx-minx,maxy-miny
ax.set_xlim(minx-.2*w,maxx+.2*w)
ax.set_ylim(miny-.2*h,maxy+.2*h)

#Set aspect ratio
ax.set_aspect(1)

#Create container for block group patches
bg_patches=[]

#For each block group...
for i,bg in enumerate(bg_list):
    #...assign a color based upon the centroid/planning area mapping...
    col=cm(amatches[i]/float(num_colors))
    #...generate a patch representing the neighborhood and throw it in the list
    bg_patches.append(PolygonPatch(bg,fc=col,ec='#555555',lw=.2,alpha=.4,zorder=1))
    
#Add the patches to the plot object en masse
ax.add_collection(PatchCollection(bg_patches,match_original=True))

#Define simple function to plot Shapely polygons
def plot_poly(ax,ob,col='#999999',lw=.5):
    x,y=ob.exterior.coords.xy
    ax.plot(x,y,color=col,linewidth=lw,solid_capstyle='round',zorder=1)

#Plot planning areas
for area in pa_list:
    plot_poly(ax,area,lw=1.5)
    

#Remove ticks
ax.set_xticks([])
ax.set_yticks([])

#Set title
plt.title('Mapping Census Block Groups to Planning Areas')
plt.legend()

#Maximize plot usage
plt.tight_layout()
```

It appears that our centroid rule does a pretty good job of staying true to the planning areas.  Disparities between the groupd are relatively minor, and modifications of the current groupings have their own downsides as well.  This is enough to proceed with the union.

## Block Group Unions

To capture the constituent block groups for each new neighborhood, we already have the building blocks required.  We simply need to collect them in one DataFrame and subset appropriately.  We can perform the same operation for all areas and throw the DataFrames in a list.

```python
#Create DF with block group info
bg_df=DataFrame({'geoid':bg_id_list,
                 'geom':bg_list,
                 'area':amatches})

#Create container for area specific subsets
area_subs=[]

#For each planning area...
for area in sorted(set(bg_df['area'])):
    #...subset by said area and throw it in the list
    area_subs.append(bg_df[bg_df['area']==area])

area_subs[0].head()
```

Now we can grab the **`Polygon`** objects from each subset, and consolidate them into a single **`Polygon`** object with the `unary_union()` method.

```python
#Create container for MultiPolygons
multi_areas=[]

#For each planning area...
for area_df in area_subs:
    #...convert the associated block group Polygons to a single Polygon and throw it in a list
    multi_areas.append(unary_union(list(area_df['geom'].values)))

multi_areas[0]
```

Let's visually confirm that we have constructed the **`Polygon`** objects we desired.  We will ultimately want to couple names to our new **`Polygon`** objects, so we should see if the lists line up as well.  The numbers come from two sources.  The black numbers are the indices of our newly constructed **`Polygon`** list.  The red numbers are the indices of the original planning area **`Polygon`** list (where our area names come from).

```python
#Generate plot object
fig,ax=plt.subplots(figsize=(16,14))

#Set scope of plot
minx,miny,maxx,maxy=MultiPolygon(multi_areas).bounds
w,h=maxx-minx,maxy-miny
ax.set_xlim(minx-.2*w,maxx+.2*w)
ax.set_ylim(miny-.2*h,maxy+.2*h)

#Set aspect ratio
ax.set_aspect(1)

#Plot planning areas
for i,marea in enumerate(multi_areas):
    plot_poly(ax,marea,lw=1.5)
#     ax.annotate(pa_name_list[i],xy=marea.centroid.coords[0],size=10)
    ax.annotate(i,xy=marea.centroid.coords[0],size=10)
    ax.annotate(i,xy=(pa_list[i].centroid.x-500,pa_list[i].centroid.y-500),size=10,color='r')
    

#Remove ticks
ax.set_xticks([])
ax.set_yticks([])

#Set title
plt.title('New Neighborhoods Built from Block Groups')
plt.legend()

#Maximize plot usage
plt.tight_layout()
```

The first five look ok, but the rest are clearly out of sync.  One way to deal with this would be to leverage spatial indexing yet again.  The most elegant solution, however, is not always required.  There are only ten areas, so a simple dictionary is pretty low cost to build.

```python
#Create Polygon matching dictionary
poly_match={0:0,
            1:1,
            2:2,
            3:3,
            4:4,
            5:9,
            6:5,
            7:6,
            8:7,
            9:8}

#Generate plot object
fig,ax=plt.subplots(figsize=(16,14))

#Set scope of plot
minx,miny,maxx,maxy=MultiPolygon(multi_areas).bounds
w,h=maxx-minx,maxy-miny
ax.set_xlim(minx-.2*w,maxx+.2*w)
ax.set_ylim(miny-.2*h,maxy+.2*h)

#Set aspect ratio
ax.set_aspect(1)

#Plot planning areas
for i,marea in enumerate(multi_areas):
    plot_poly(ax,marea,lw=1.5)
    centroid=(marea.centroid.x-1000,marea.centroid.y-100)
    ax.annotate(pa_name_list[poly_match[i]].replace(' ','\n'),xy=centroid,size=7)

#Remove ticks
ax.set_xticks([])
ax.set_yticks([])

#Set title
plt.title('New Neighborhoods Built from Block Groups')
plt.legend()

#Maximize plot usage
plt.tight_layout()
```

## Writing Output to Disk

It looks like our labels are in order.  We now have the necessary components for our new shapefile, and we can write to disk. 

```python
#Define the schema for the new shapefile
schema={'properties':{'NAME':'str'},
        'geometry':'Polygon'}

#Open a new shapefile...
with fiona.open(dc_dir+'CensusBG_Areas.shp','w','ESRI Shapefile',schema,bgroups.crs) as output:
    #...and for each new neighborhood...
    for i,marea in enumerate(multi_areas):
        #...write the geom and name to the shapefile
        output.write({'properties':{'NAME':pa_name_list[poly_match[i]]},
                      'geometry':mapping(marea)})
```

We have created a very simple shapefile with 10 planning areas built from Census block groups in the District.  The [output](http://cdb.io/1fd58W2) can be viewed interactively via [CartoDB](http://cartodb.com/).
