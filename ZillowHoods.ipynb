# Neighborhoods as Defined by Zillow

It appears that [Zillow](http://www.zillow.com/) has made the [shapefiles](http://www.zillow.com/howto/api/neighborhood-boundaries.htm) they use to categorize home locations available to the public.  This script is just intended to get a view of these neighborhoods.

```python
import numpy as np
import fiona 
from shapely.geometry import shape,Polygon,MultiPolygon
from descartes import PolygonPatch
from matplotlib.collections import PatchCollection
```

Let's read in the data.

```python
#Establish data directory
data_dir='/home/choct155/dissertation/MiscData/'

#Open neighborhood shapefile
zillow=fiona.open(data_dir+'ZillowNeighborhoods-DC.shp')

#Initiate iterator
nbhd=zillow.next()

#Create container for features and names
nbhd_list=[]
nbhd_name_list=[]

#For each feature...
for nbhd in zillow:
    #...throw the feature in the list...
    nbhd_list.append(shape(nbhd['geometry']))
    #...and its name into another list
    nbhd_name_list.append(nbhd['properties']['NAME'])

nbhd
```

As we can see, it's an expectedly minimal set, but it does provide a name handle.

```python
print 'THERE ARE',len(nbhd_list),'NEIGHBORHOODS IN THE FILE'
```

It carries about half the neighborhoods present in [DC's assessment neighborhood](http://dcatlas.dcgis.dc.gov/metadata/AsNbhdPly.html) shapefile (which holds 71).  Let's quickly validate to ensure all objects in the set are of type `Polygon`.

```python
nbhd_dict=dict(zip(nbhd_name_list,nbhd_list))

[x for x in nbhd_dict.items() if x[1].geom_type != 'Polygon']
```

Looks like we are good there, so let's take a look at a plot.

```python
#Generate plot object
fig = plt.figure(figsize=(16,16))
ax = fig.add_subplot(111)

#Get color map
cm = plt.get_cmap('RdBu')
num_colours = len(nbhd_list)

#Set scope of the plot
minx, miny, maxx, maxy = MultiPolygon(nbhd_list).bounds
w, h = maxx - minx, maxy - miny
ax.set_xlim(minx - 0.2 * w, maxx + 0.2 * w)
ax.set_ylim(miny - 0.2 * h, maxy + 0.2 * h)

#Set aspect ratio
ax.set_aspect(1.2)

#Create container for patches (neighborhoods in this context) and locations
patches = []
patch_loc=[]

#Define corresponding lists for shapely objects and names
nbhd_objects=[x[1] for x in sorted(nbhd_dict.items())]
nbhd_names=[x[0] for x in sorted(nbhd_dict.items())]

#For each neighborhood...
for idx, p in enumerate(nbhd_objects):
    #...assign a color based upon index position in the MP object...
    colour = cm(1. * idx / num_colours)
    #...generate a patch representing the neighborhood and throw it in the list...
    patches.append(PolygonPatch(p, fc=colour, ec='#555555', lw=0.2, alpha=.4, zorder=1,label=nbhd_names[idx]))
    #...and capture the patch location
    patch_loc.append((p.centroid.x-.005,p.centroid.y))

#Add the patches to the plot object en masse
ax.add_collection(PatchCollection(patches, match_original=True))

#Add labels
for i,loc in enumerate(patch_loc):
    ax.annotate(nbhd_names[i],xy=loc,size=5)

#Remove ticks
ax.set_xticks([])
ax.set_yticks([])

#Set title
plt.title("Zillow's Mapping of Neighborhoods in DC",fontsize=22)
plt.legend()

#Maximize plot usage by minimizing margins
plt.tight_layout()

plt.savefig('ZillowNbhdDC.png',dpi=250)
```
